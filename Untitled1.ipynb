```python
# === 1) Instalar dependencias ===
!pip -q install fastapi uvicorn pyngrok "python-multipart" requests

# === 2) Configurar variables ===
import os, json, threading, time, requests
from pyngrok import ngrok, conf
from fastapi import FastAPI, Request
from fastapi.responses import PlainTextResponse, JSONResponse

# üëâ Rellena estos 3 valores con lo que ves en tu panel de WhatsApp Cloud API
WHATSAPP_TOKEN      = os.environ.get("WHATSAPP_TOKEN"     , "PASTE_TU_ACCESS_TOKEN")
PHONE_NUMBER_ID     = os.environ.get("PHONE_NUMBER_ID"    , "PASTE_TU_PHONE_NUMBER_ID")
VERIFY_TOKEN        = os.environ.get("VERIFY_TOKEN"        , "orbita_verify_123")  # el que quieras, lo mismo se pone en Meta

# üëâ (Opcional) tu Authtoken de ngrok (si no lo pones, a veces ngrok no abre t√∫nel)
NGROK_TOKEN         = os.environ.get("NGROK_TOKEN"         , "PASTE_TU_NGROK_AUTHTOKEN")

# === 3) App FastAPI ===
app = FastAPI(title="Orbita WhatsApp Webhook")

# --- 3.1 Verificaci√≥n de webhook (Meta env√≠a hub.* por GET) ---
@app.get("/webhook", response_class=PlainTextResponse)
async def verify(mode: str = None, hub_mode: str = None, hub_challenge: str = None, hub_verify_token: str = None,
                 challenge: str = None, **query):
    # Meta env√≠a: hub.mode, hub.verify_token, hub.challenge
    token = hub_verify_token or query.get("hub.verify_token")
    ch    = hub_challenge    or query.get("hub.challenge") or challenge
    if token == VERIFY_TOKEN and ch:
        return ch
    return PlainTextResponse("Error: verify token inv√°lido", status_code=403)

# --- 3.2 Recepci√≥n de mensajes (POST) ---
@app.post("/webhook")
async def incoming(req: Request):
    payload = await req.json()
    # Log plano para depurar:
    print("== Webhook RX ==")
    print(json.dumps(payload, indent=2, ensure_ascii=False))

    # Extraer mensaje si existe:
    try:
        entry = payload["entry"][0]
        changes = entry["changes"][0]
        value = changes["value"]
        contacts = value.get("contacts", [])
        messages = value.get("messages", [])

        if messages:
            msg = messages[0]
            from_wa = msg["from"]              # e.g. "54911XXXXXXX"
            text = None

            if msg["type"] == "text":
                text = msg["text"]["body"].strip()

            # Respuesta simple (puedes poner tu flujo/men√∫ aqu√≠):
            if text:
                reply = (
                    "Hola üëã, te escribe *√ìrbita*.\n"
                    "Eleg√≠ una opci√≥n:\n"
                    "A) An√°lisis comercial puntual\n"
                    "B) An√°lisis completo HTML\n"
                    "C) Asesor√≠a digital con herramientas\n\n"
                    "Escrib√≠ A, B o C."
                )
                if text.lower() == "a":
                    reply = "Opci√≥n A ‚úÖ\n¬øMedio de pago, zona espec√≠fica, productos o *otros*?"
                elif text.lower() == "b":
                    reply = "Opci√≥n B ‚úÖ\nElijo: (a) productos en stock, (b) marketing y ventas, (c) an√°lisis geogr√°fico."
                elif text.lower() == "c":
                    reply = "Opci√≥n C ‚úÖ\n(a) ventas/productos, (b) medios de pago, (c) an√°lisis demogr√°fico, (d) incorporar estrategias + asesor."
                send_text(from_wa, reply)

    except Exception as e:
        print("Parse error:", e)

    return JSONResponse({"status":"ok"})

# --- 3.3 Enviar mensaje por WhatsApp ---
def send_text(to_number: str, body: str):
    url = f"https://graph.facebook.com/v20.0/{PHONE_NUMBER_ID}/messages"
    headers = {
        "Authorization": f"Bearer {WHATSAPP_TOKEN}",
        "Content-Type": "application/json"
    }
    data = {
        "messaging_product": "whatsapp",
        "to": to_number,
        "type": "text",
        "text": {"preview_url": False, "body": body}
    }
    r = requests.post(url, headers=headers, json=data)
    print("== TX ==", r.status_code, r.text)

# === 4) Levantar Uvicorn + Ngrok y mostrar la URL ===
def run_uvicorn():
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)

# Ngrok
ngrok.kill()
if NGROK_TOKEN and NGROK_TOKEN != "PASTE_TU_NGROK_AUTHTOKEN":
    conf.get_default().auth_token = NGROK_TOKEN
public_url = ngrok.connect(8000, bind_tls=True).public_url

# Iniciar servidor en hilo aparte
threading.Thread(target=run_uvicorn, daemon=True).start()
time.sleep(2)

print("‚úÖ Server OK")
print("üåê URL p√∫blica:", public_url)
print("üìå Callback URL para Meta:", public_url + "/webhook")
print("üîë VERIFY_TOKEN (pon√© exactamente este en Meta):", VERIFY_TOKEN)
print("\n‚û° En WhatsApp Cloud API > Webhooks: pega la Callback URL y este VERIFY_TOKEN.")
print("‚û° Luego suscribite al campo 'messages' y prob√° enviando un WhatsApp al n√∫mero de prueba.")
```
